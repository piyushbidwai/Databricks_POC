```python
%sql
CREATE OR REPLACE FUNCTION training.piyush.mask_email_data(EmailAddress STRING)
RETURN

  CASE 
      WHEN current_user() = 'piyush.bidwai@inferenz.ai' OR is_account_group_member('DataEngineer') THEN EmailAddress
       ELSE CONCAT(SUBSTRING(EmailAddress, 1, 2), '*****@*****', SUBSTRING(EmailAddress, LENGTH(EmailAddress)-1, 2))
  END
```
