```python
from pyspark.sql.functions import *
# from pyspark.sql.windows import Window
```

```python
salesdf = spark.read.table('training.piyush.sales_fact_table')
salesdf.display()
```

```python
salesdf.printSchema()
```

**KPI - Sales_per_year_per_Month**

```python
salesdfv1 = salesdf.withColumn('Year', year(to_date(salesdf['date_key'], 'yyyy-MM-DD')))\
                      .withColumn('Month', month(to_date(salesdf['date_key'], 'yyyy-MM-DD')))


sales_per_yr_per_month = salesdfv1.groupBy('Year', 'Month').agg(count('sales_order_id').alias('orders_per_month')).orderBy(col('orders_per_month').desc())


sales_per_yr_per_month.display()

sales_per_yr_per_month.write.mode('overwrite').saveAsTable('training.piyush.kpi1_sales_per_yr_per_month')
```

**KPI - Number of Orders per customer**

```python
order_per_cust = salesdf.groupBy('customer_key').agg(count('sales_order_id').alias('orders_per_customer'))
order_per_cust.display()

order_per_cust.write.mode('overwrite').saveAsTable('training.piyush.kpi2_orders_per_customer')
```

**KPI - Average Unit Price per Product**

```python

Avgper_product = salesdf.groupBy('product_key').agg(avg('UnitPrice').alias('avg_UnitPrice'))
Avgper_product.show()

Avgper_product.write.mode('overwrite').saveAsTable('training.piyush.kpi3_avg_unitprice_per_product')
```

**KPI - Average Value spent per customer for all his products listed**

```python
Avgvalue_spent = salesdf.groupBy('customer_key').agg(avg('UnitPrice').alias('Avgvalue_spent'))
Avgvalue_spent.show()

Avgvalue_spent.write.mode('overwrite').saveAsTable('training.piyush.kpi4_Avgvalue_spent')
```

**KPI - Top 10 customers having highest nummber of products**

```python
topcustomers = salesdf.groupBy('customer_key').agg(count('product_key').alias('No_of_products'))
topcustomers = topcustomers.orderBy(col('No_of_products'), ascending = False).limit(10)
topcustomers.show()

topcustomers.write.mode('overwrite').saveAsTable('training.piyush.kpi5_topcustomers')
```
